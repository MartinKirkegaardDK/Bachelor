```python
import matplotlib.pyplot as plt

list1 = [0.44, 0.23, 0.23, 0.06]
list2 = [0.58, 0.44, 0.44, 0.47]

fig, ax = plt.subplots()

bar_width = 0.4
x_pos = [i for i in range(len(list1))]
x_pos2 = [x + bar_width for x in x_pos]

ax.bar(x_pos, list1, color='blue', label='Linear regression', width=bar_width)
ax.bar(x_pos2, list2, color='red', label='Random Forest', width=bar_width)

ax.set_xlabel('Distance Metrics')
ax.set_ylabel('R-squared score')
ax.set_title('R-squared score of Different Models using Distance Metrics')
ax.set_xticks([i + bar_width/2 for i in x_pos])
ax.set_xticklabels(['Cosine', 'Eucledian', 'Manhatten', 'Heterogeneous'])
ax.legend()

plt.show()
plt.savefig("pictures_for_rapport/r2_comp_dist")
```
